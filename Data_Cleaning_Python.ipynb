### Creating function rm_tags to remove html tags from the data

```python
def rm_tags(raw_html):
   clean=BeautifulSoup(raw_html, "lxml").text
   return clean
```

### Importing Regular expression module

```python
import re
import string
```

### Defining function rm_punc to remove punctuations from the data

```python
def rm_punc(word):
   pattern = r'[' + string.punctuation + ']'
   return re.sub(pattern, '', word)
```

```python
df["Body"]=df["Body"].apply(rm_punc)
```

```python
df1["Body"]=df1["Body"].apply(rm_punc)
df2["Body"]=df2["Body"].apply(rm_punc)
df3["Body"]=df3["Body"].apply(rm_punc)
```

```python
df["Title"]=df["Title"].apply(rm_punc)
```

```python
df1["Title"]=df1["Title"].apply(rm_punc)
df2["Title"]=df2["Title"].apply(rm_punc)
df3["Title"]=df3["Title"].apply(rm_punc)
```

### Replacing \n new line, \t tab, \r carriagereturn, \b backspace, \f form feed

```python
df = df.replace(r'\n',' ', regex=True)
df = df.replace(r'\t',' ', regex=True)
df = df.replace(r'\r',' ', regex=True)
df = df.replace(r'\b',' ', regex=True)
df = df.replace(r'\f',' ', regex=True)
```

```python
df1 = df1.replace(r'\n',' ', regex=True)
df1 = df1.replace(r'\t',' ', regex=True)
df1 = df1.replace(r'\r',' ', regex=True)
df1 = df1.replace(r'\b',' ', regex=True)
df1 = df1.replace(r'\f',' ', regex=True)
```

```python
df2 = df2.replace(r'\n',' ', regex=True)
df2 = df2.replace(r'\t',' ', regex=True)
df2 = df2.replace(r'\r',' ', regex=True)
df2 = df2.replace(r'\b',' ', regex=True)
df2 = df2.replace(r'\f',' ', regex=True)
```

```python
df3 = df3.replace(r'\n',' ', regex=True)
df3 = df3.replace(r'\t',' ', regex=True)
df3 = df3.replace(r'\r',' ', regex=True)
df3 = df3.replace(r'\b',' ', regex=True)
df3 = df3.replace(r'\f',' ', regex=True)
```
